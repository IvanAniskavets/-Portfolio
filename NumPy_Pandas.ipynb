На сайт заходит очередной посетитель, о покупках которого известно следующее:

Найдите самого похожего пользователя. Т. е. посчитайте косинусное сходство между этим пользователем и всеми пользователями из массива user_stats

### Решение

```python
dists = []
lines = len(users_stats)
i = 0
for i in range(lines):
    """
    Подсчет косинуса угла между векторами a, b по их координатам
    """
    # длины векторов
    aLength = np.linalg.norm(users_stats[i])
    bLength = np.linalg.norm(next_user_stats)
    cosine = round(np.dot(users_stats[i], next_user_stats) / (aLength * bLength), 2)
    dists.append({"index_user": i, "distance": float(cosine), "vector": users_stats[i]}) 
    """
    создание результирующщего словаря из которого будет найдено max значение косинусного сходства между user'ами
    """
    i += 1
```

### Задание 8

Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

### Решение

```python
# открываем и читаем таблицу с рейтингами

ratings = pd.read_csv('../ml-25m/ratings.csv')
ratings.head()
```

```python
# считаем средний рейтинг для каждого фильма и формируем таблицу со средним рейтингом

average_rating = ratings[['movieId','rating']].groupby('movieId').mean().round(2)
average_rating.head()
```

```python
# функция присваивающая классификацию фильма на основе его рейтинг

def class_movie(rating):
    if rating <= 2:
        return 'Низкий рейтинг'
    elif rating > 2 and rating <= 4:
        return  'Средний рейтинг'
    elif rating == 4.5 or rating == 5:
        return 'Высокий рейтинг'
```

```python
# применяем функцию к датафрейму со средним рейтигом фильмов

average_rating['class'] = average_rating['rating'].apply(class_movie)
average_rating.head()
```

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

### Решение

```python
# читаем наш файл

keywords = pd.read_csv('C:/Netology/Python_11_pandas_functions_groupby/keywords.csv')
keywords.head()
```

```python
# функция которая ищет название города в запросе пользователя

def geo_classifier(keyword):
    for key, value in geo_data.items(): #используя цикл для того чтобы добраться до названия города и присвоеть его переменной City
        for city in value:
            if city in keyword.lower().split(' '):
                return key        
    return 'undefined'
```

```python
# применяем функцию к датафрейму с поисковыми запросами

keywords['region'] = keywords['keyword'].apply(geo_classifier)
keywords.sample(10)
```

### Проверка количества поисковых запросов содержащих города регионов

```python
# делаем проверку чтобы найти поисковые запросы с указанием городов регионов

keywords.groupby('region').count().reset_index()[['region', 'keyword']]
```

для каждой строки пройдите по всем годам списка years

```python
# список годов от 1950 по 2010

years = list(range(1950, 2011))
```

```python
# функция ищет год в названии фильма

def production_year(title):
    for year in years:  #через цикл проверяем каждый год на наличие в названии фильма
        if str(year) in title:
            return int(year)
    return 1900
```

```python
# применяем функцию к датафрейму

tab_movies['year'] = tab_movies['title'].apply(production_year)
tab_movies.head()
```

```python
# фильтруем нужные нам столбцы с учёт периода с 1950 по 2010 года

movies_year = tab_movies[['movieId', 'title', 'year']].query('year >= 1950 and year <= 2010')
movies_year.head()
```
