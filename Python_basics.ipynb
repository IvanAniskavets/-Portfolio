```python
def name_owner():
    user_input = input('Введите номер документа: ')
    for document in documents:
        number = document['number']
        name = document['name']
        if user_input == number:
            name = name
            break
        else:
            name = None
    if name != None:
        print(f'Владелец документа: {name}')
    else:
        print('Владелец документа не найден')

name_owner()
```

```python
def number_directory():
    user_input = input('Введите номер документа: ')
    for key, value in directories.items():
        if user_input in value:
            key = key
            break
        else:
            key = None
    if key != None:
        print(f'Документ хранится на полке: {key}')
    else:
        print('Документ не найден или некорретный номер документа')
        
number_directory()
```

```python
def info_total():
    for document in documents:
                type_doc = document['type']
                doc_num = document['number']
                name_own =  document['name']
                for key, value in directories.items():
                    if document['number'] in value:
                        key = key
                        print(f'№: {doc_num}, тип: {type_doc}, владелец: {name_own}, полка хранения: {key}')
                
info_total()
```

```python
def get_shelf():
    user_input = input('Введите номер полки: ')
    for key in directories.keys():
        key
    if user_input not in directories:
        directories[user_input] = []
        print('Полка добавлена. Текущий перечень полок:', ', '.join(directories.keys()), '.')        
    else:
        print('Такая полка уже существует. Текущий перечень полок:', ', '.join(directories.keys()), '.')
    
get_shelf()
```

```python
def delete_shelf():
    user_input = input('Введите номер полки: ')
    if user_input in list(directories.keys()) and directories[user_input] == [] :
        del(directories[user_input])
        print('Полка удалена.Текущий перечень полок:', ', '.join(directories.keys()), '.')
    elif user_input in list(directories.keys()) and directories[user_input] != [] :
        print('На полке есть документы, удалите их перед удалением полки.Текущий перечень полок:', ', '.join(directories.keys()), '.')
    else:
        print('Такой полки не существует. Текущий перечень полок:', ', '.join(directories.keys()), '.')
    
        
delete_shelf()
```

```python
def get_document():
    user_input = input('Введите номер документа: ')
    number_doc = user_input
    user_input_2 = input('Введите тип документа: ')
    type_doc = user_input_2
    user_input_3 = input('Введите владельца документа: ')
    owner_doc = user_input_3
    user_input_key = input('Введите полку для хранения: ')
    key = user_input_key
    if key in directories.keys():
        res = {}
        res['type'] = type_doc
        res['number'] = number_doc
        res['name'] = owner_doc
        documents.append(res)
        directories[key].append(number_doc)
        print('Документ добавлен. Текущий список документов:')
        info_total()
    else:
        print('Такой полки не существует. Добавьте полку командой as.')
        info_total()
        
get_document()
```

```python
def delete_document():
    user_input = input('Введите номер документа: ')
    for value in documents:
        if user_input in list(value.values()):
            documents.remove(value)
            print('Документ удален.')
            print('Текущий список документов:')
            info_total()
            for key, value in directories.items():
                if user_input in value:
                    value.remove(user_input)
            return
    print('Документ не найден в базе.')
    print('Текущий список документов:')
    info_total()
    return
                
delete_document()
```

```python
def move_doc():
    doc_number = input('Введите номер документа: ')
    number_directory = input('Введите номер полки:')
    if number_directory not in directories:
        print('Такой полки не существует. Текущий перечень полок:', ', '.join(directories.keys()), '.')
        return
    for key, value in directories.items():
        if doc_number in value:
            directories[number_directory] += [doc_number]
            value.remove(doc_number)
            print('Документ перемещен')
            print('Текущий список документов:')
            info_total()
            return
    print('Документ не найден в базе.')
    print('Текущий список документов:')
    info_total()
    return

move_doc()
```

```python
# Код итоговая главнай функции

def main():
    print('Программа запущена и готова к работе')
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            print('Результат:')
            name_owner()
        elif user_input == 's':
            print('Результат:')
            number_directory()
        elif user_input == 'l':
            print('Результат:')
            info_total()
        elif user_input == 'ads':
            print('Результат:')
            get_shelf()
        elif user_input == 'ds':
            print('Результат:')
            delete_shelf()
        elif user_input == 'ad':
            print('Результат:')
            get_document()
        elif user_input == 'd':
            print('Результат:')
            delete_document()
        elif user_input == 'm':
            print('Результат:')
            move_doc()
        elif user_input == 'q':
            print('Работа завершена.')
            break
        else:
            print('Введена неверная команда.')
```

### Задание 2
Добавьте в класс Rate параметр diff (со значениями True или False), который в случае значения True в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением. Считайте, self.diff будет принимать значение True только при возврате значения курса. При отображении всей информации о валюте он не используется.

### Решение

```python
class Rate:
    def __init__(self, diff=True, format_='value'):
        self.format = format_
        self.diff = diff
    
    def exchange_rates(self):
        """
        Возвращает ответ сервиса с информацией о валютах в виде:
        
        {
            'AMD': {
                'CharCode': 'AMD',
                'ID': 'R01060',
                'Name': 'Армянских драмов',
                'Nominal': 100,
                'NumCode': '051',
                'Previous': 14.103,
                'Value': 14.0879
                },
            ...
        }
        """
        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
    
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate('full').make_format('EUR')
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        
        Rate('value').make_format('EUR')
        79.4966
        """
        response = self.exchange_rates()
        
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            # Задание № 2 добавление параметра diff
            if self.format == 'value':
                if self.diff == True: #условие значения для параметра diff
                    return round(response[currency]['Previous'] - response[currency]['Value'], 4) # разница между предудущим и текущим значением курса валюьы
                else:
                    return response[currency]['Value']
        
        return 'Error'
    
    def eur(self):
        """Возвращает курс евро на сегодня в формате self.format"""
        return self.make_format('EUR')
    
    def usd(self):
        """Возвращает курс доллара на сегодня в формате self.format"""
        return self.make_format('USD')
    
    def brl(self):
        """Возвращает курс бразильского реала на сегодня в формате self.format"""
        return self.make_format('BRL')
    
```

```python
class Employee:
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        
        self.grade = 1
    
    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade += 1
    
    def publish_grade(self):
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)
    
    def check_if_it_is_time_for_upgrade(self):
        pass
```

```python
class Developer(Employee):
    def __init__(self, name, seniority):
        super().__init__(name, seniority)
    
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        # пока считаем, что все разработчики проходят аккредитацию
        self.seniority += 1
        
        # условие повышения сотрудника из презентации
        if self.seniority % 5 == 0:
            self.grade_up()
        
        # публикация результатов
        return self.publish_grade()
```

```python
class Designer(Employee):
    def __init__(self, name, seniority, awards):
        super().__init__(name, seniority)
        self.awards = awards
    
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        # пока считаем, что все разработчики проходят аккредитацию
      
        
        # условие повышения сотрудника при наличии международной премии
        if self.awards != 0:
            for i in range(self.awards*2):
                self.seniority += 1   # повышаем баллы сотрудника n раз, где n = количество премий * количество баллов за одну премию
                if self.seniority % 7 == 0: # проверяем на кратность 7 по условию повышения грейда
                    self.grade_up()
                    self.awards = 0
                else:
                    self.awards = 0   #обнуляем количество премий чтобы их учесть разово на старте каррьеры 
        else:
            self.seniority += 1
            if self.seniority % 7 == 0:
                self.grade_up()
        
            
        # публикация результатов
        return self.publish_grade()
```

Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.

```python
def valid_number_phone(phone):
    pattern = r'^\+?[78][-\(]?\s?[9]\d{2}\)?-?\s?\d{3}-?\s?\d{2}-?\s?\d{2}$'
    if re.findall(pattern, phone):
        number_phone = re.findall(pattern, phone)
        res = []
        for el_1 in number_phone:
            for el_2 in el_1:
                try:
                    el_2 == int(el_2)
                    res.append(el_2)
                except ValueError:
                    pass
        my_str = ''.join(res)    
        print (f'Результат: +7-{my_str[1:4]}-{my_str[4:7]}-{my_str[7:9]}-{my_str[-2:]}')
    else:
        print('Номер не валиден')
```
